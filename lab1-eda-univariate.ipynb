**Business Question**: How are products priced on Amazon UK, and are there specific price points or ranges that are more common?

1. **Measures of Centrality**:
    - Calculate the mean, median, and mode for the `price` of products.
    - What's the average price point of products listed? How does this compare with the most common price point (mode)?

2. **Measures of Dispersion**:
    - Determine the variance, standard deviation, range, and interquartile range for product `price`.
    - How varied are the product prices? Are there any indicators of a significant spread in prices?

3. **Visualizations**:

```python
""" 2.1 Measures of Centrality
    - Calculate the mean, median, and mode for the `price` of products.
    
"""
```

```python
df = pd.read_csv('amz_uk_price_prediction_dataset.csv')
df.head()
```

```python
# 2.1.1 Calculate the mean, median, and mode for the `price` of products.

df['price'].describe()
```

```python
mean_price = df['price'].mean()
print("the valeo of mean price is : ",mean_price)

median_price = df['price'].median()
print("the valeo of median price is : ",median_price)

mode_price = df['price'].mode()
print("the valeo of mode price is : ",mode_price)
```

```python
# 2.1.2 What's the average price point of products listed? How does this compare with the most common price point (mode)?
```

```python
average_price = df['price'].mean()
print("the average price point of the products listed is : ", average_price, "$")
```

```python
# How does this compare with the most common price point (mode)?

mode_price = df['price'].mode()[0]   # [0] is used to select the first value

print("The most common price point of products listed is: ", mode_price, "$")

# The most common price point of products listed of 9.99€ is a lot lower then the average price (89.24€)
```

2.2 **Measures of Dispersion**:

    2.2.1 - Determine the variance, standard deviation, range, and interquartile range for product `price`.
    2.2.2 - How varied are the product prices? Are there any indicators of a significant spread in prices?
    

```python
#    2.2.1 - Determine the variance, standard deviation, range, and interquartile range for product `price`.
```

```python
# Determine the variance
print(f' The value of the variance is: {df['price'].var()}')

# Determine the standard deviation
print(f' The value of the standard deviation is: {df['price'].std()}')

# Determine the range of values
print(f' The range of values is: {df['price'].max()-df['price'].min()}')

# Determine the interquartile range
print(f' the interquartile range is: {df['price'].quantile(0.75)-df['price'].quantile(0.25)}')
```

```python
# 2.2.2 - How varied are the product prices? Are there any indicators of a significant spread in prices?

print(f'the value of the skewness of prices is : ',{df['price'].skew()})
print(f'the value of the kurtosis of prices is : ',{df['price'].kurtosis()})

```

```python

# Skewness: 
# The positive value of skewness (27.49) for the price indicates that the distribution is right-skewed, indicating a significant number of products that have a higher price,
# and we can consider them as outliers.

# Kurtosis: 
# The kurtosis value is 4241.62 and its greater than 3, which indicates that the price distribution has big and havy tails.



"""
(with some help)
The variance of product prices is indicating that the prices are spread out from the mean by a considerable amount, refering to a significant level of variability in product prices.

The standard deviation indicates a wide spread in prices.

The indicates a very big significant variability in the prices of products.

The interquartile range (IQR) of 36.0 indicates that the middle 50% of the prices are spread over a relatively narrow range,
but the overall distribution still shows a significant spread.

"""
```
